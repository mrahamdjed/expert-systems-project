```python
rules={
    "cat":
    [
        {"condition":["four_legs","meows"],"conclusion":"cat","additional":"Domesticated, Known for hunting rodents"},
        {"condition":["four_legs","rectangle_claws"],"conclusion":"cat","additional":"Domesticated, Known for hunting rodents"},
    ],
    "persian_cat":
    [
        {"condition":["cat","long_fur","small_size"],"conclusion":"persian_cat","additional":"Domesticated, Known for its elegant appearance"},
    ],
    "bird":
    [
        {"condition":["feathers","can_fly"],"conclusion":"bird","additional":"Lives in trees, Lays eggs"}
    ],
}

facts=["four_legs","meows","long_fur","small_size"]
```

```python


goals=["persian_cat","cat","bird"]

def Backward_Chaining(goals,facts,rules):
    for goal in goals:
        if goal in facts:
            return True
        
    for goal in goals:
        for rule in rules.get(goal):
            if all(condition in facts for condition in rule["condition"]):
                return rule
        
    return False

result=Backward_Chaining(goals,facts,rules)
print("Animal :",result["conclusion"],"\nAdditional Information :",result["additional"])
```
